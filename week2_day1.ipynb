````python
'''1. Prime Number Finder
 
   * Function that returns all prime numbers up to `n`.
   * Input: `n = 20`
   * Output: `[2, 3, 5, 7, 11, 13, 17, 19]`'''
def find_primes(n):
    primes = []
    for num in range(2, n + 1):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
            if is_prime:
                primes.append(num)
    return primes
n = 20  
print(find_primes(n))
````

````python
'''2. Palindrome Checker
 
   * Check if a string is palindrome (ignoring spaces & case).
   * Input: `"Never odd or even"`
   * Output: `True`'''
def is_palindrome(s):
    cleaned = ''.join(c.lower() for c in s if c.isalnum())
    return cleaned == cleaned[::-1]

s = input("Enter a string: ")
print(is_palindrome(s))
````

````python
'''Anagram Checker
 
   * Function checks if two words are anagrams.
   * Input: `"listen", "silent"`
   * Output: `True`'''
def anagrams(word1,word2):
    return sorted(word1).lower() == sorted(word2).lower()
word1 = input("Enter first word: ")
word2 = input("Enter second word: ")
print(anagrams(word1,word2))
````

````python
'''Count Vowels & Consonants
 
   * Return count of vowels & consonants.
   * Input: `"hello"`
   * Output: `Vowels=2, Consonants=3`'''
def count_vowels_consonants(s):
    vowels = 'aeiouAEIOU'
    v_count = sum(1 for char in s if char in vowels)
    c_count = sum(1 for char in s if char.isalpha() and char not in vowels)
    return v_count, c_count
s=input("Enter a string: ")
v,c=count_vowels_consonants(s)
print(f"Vowels={v}, Consonants={c}")
````

````python
'''Longest Word Finder
 
* Return longest word in a sentence.
* Input: `"Python is amazing"`
* Output: `"amazing"`'''
def longest_word(sentence):
    words = sentence.split()
    return max(words, key=len)
sentence = input("Enter a sentence: ")
print(longest_word(sentence))
````

````python
'''Random Password Generator (`random`, `string`)
 
* Input: `length=6`
* Output: `"aB9@xQ"`
''' 
  
import random
import string
def random_pswd(n):
    len=6
    if n<4:
        return "Length should be at least 4"    
    characters = string.ascii_letters + string.digits + string.punctuation
    return ''.join(random.choice(characters) for i in range(n))
n=int(input("Enter length of password: "))
print(random_pswd(n))
    
````

````python
'''Simple Calculator with `math`
 
* Perform sqrt, factorial, sin, cos.
* Input: `sqrt(25)`
* Output: `5.0`'''
import math
def calculator():
    print("Select operation:")
    print("1. Square Root")
    print("2. Factorial")
    print("3. Sine")
    print("4. Cosine")
    choice = input("Enter choice(1/2/3/4): ")
    num = float(input("Enter number: "))
    match choice:
        case '1':
            print(f"Square Root of {num} is {math.sqrt(num)}")
        case '2':
            print(f"Factorial of {int(num)} is {math.factorial(int(num))}")
        case '3':
            print(f"Sine of {num} is {math.sin(num)}")
        case '4':
            print(f"Cosine of {num} is {math.cos(num)}")
        case _:
            print("Invalid input")
calculator()
            
````

````python
'''Timer Program (`time`)
 
* Input: `countdown(5)`
* Output:
 
  ```
  5  
  4  
  3  
  2  
  1  
  Time’s up!
  ```'''
import time
def count_time(n):
    while n>0:
        print(n)
        time.sleep(1)
        n-=1
    print("Time's up!")
n=int(input("Enter countdown time in seconds: "))
count_time(n)
  
````

````python
'''. Bank Account Class
 
Attributes:
 
* `balance` → stores the current money in the account
* `owner` (optional) → name of account holder
 
Methods:
 
* `deposit(amount)` → adds money to balance
* `withdraw(amount)` → subtracts money if enough funds
* `get_balance()` → returns balance
 
Example:
 
```python
acc = BankAccount(100)   # initial balance = 100
acc.deposit(50)          # balance = 150
acc.withdraw(30)         # balance = 120
print(acc.get_balance())
```
 
Output:
 
```
Balance: 120
```'''
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew: {amount}")
        else:
            print("Insufficient funds or invalid amount")

    def get_balance(self):
        return self.balance

acc = BankAccount("John", 100)   
acc.deposit(50)          
acc.withdraw(30)         
print(f"Balance: {acc.get_balance()}")
````

````python
''' Student Class with Average

 

Attributes:

 

* `name` → student name

* `marks` → list of student marks `[90, 80, 85]`

 

Methods:

 

* `get_average()` → returns average marks

* `add_mark(mark)` → add new score

* `get_highest()` → return highest mark

* `get_lowest()` → return lowest mark

 

Example:

 

```python

s = Student("Tom", [90, 80, 85])

print(s.get_average())   # (90+80+85)/3 = 85.0

s.add_mark(95)           # marks = [90,80,85,95]

print(s.get_highest())   # 95

```

 

Output:

 

```

85.0

95

'''
class Student:
    def __init__(self, name, marks=[]):
        self.name = name
        self.marks = marks

    def get_average(self):
        return sum(self.marks) / len(self.marks) if self.marks else 0

    def add_mark(self, mark):
        self.marks.append(mark)

    def get_highest(self):
        return max(self.marks) if self.marks else None

    def get_lowest(self):
        return min(self.marks) if self.marks else None
    
s = Student("Tom", [90, 80, 85]) 
print(s.get_average())  
s.add_mark(95)               
print(s.get_highest())      
print(s.get_lowest())    
print(s.marks)
````

````python
'''Employee Bonus System

 

Attributes:

 

* `name` → employee name

* `salary` → base salary

* `role` → role of employee (e.g., Manager, Developer, Intern)

 

Methods:

 

* `bonus()` → calculates bonus depending on role:

 

  * Manager → 20% of salary

  * Developer → 10% of salary

  * Intern → 5% of salary

 

Example:

 

```python

m = Manager("Alice", 50000)

print(m.bonus())

```

 

Output:

 

```

10000

```

 

---'''
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def bonus(self):
        return 0
class Manager(Employee):
    def bonus(self):
        return self.salary * 0.20
class Developer(Employee):
    def bonus(self):
        return self.salary * 0.10
class Intern(Employee):
    def bonus(self):
        return self.salary * 0.05
# Example usage
m = Manager("Alice", 50000)
print(m.bonus())  
d = Developer("Bob", 40000)
print(d.bonus())  
i = Intern("Charlie", 20000)    
print(i.bonus())  
e = Employee("Eve", 30000)
print(e.bonus()) 
````

````python
'''Shape Area Calculator

 

Base Class (Shape):

 

*Attribute: `name`

*Method: `area()` → to be overridden by subclasses

 

Subclasses:

 

* `Circle(radius)` → area = π × r²

* `Rectangle(width, height)` → area = width × height

 

Example:

 

```python

c = Circle(7)

print(c.area())

r = Rectangle(4, 5)

print(r.area())

```

 

Output:

 

```

153.94

20

```'''
import math
class Shape:
    def __init__(self, name):
        self.name = name

    def area(self):
        pass
class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius

    def area(self):
        return round(math.pi * self.radius ** 2, 2)
class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__("Rectangle")
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

c = Circle(7)
print(c.area())  
r = Rectangle(4, 5)
print(r.area())  
````

````python
''' E-commerce Order Class

 

Attributes:

 

* `items` → dictionary of items `{name: price}`

 

Methods:

 

* `add(item, price)` → adds an item to order

* `remove(item)` → removes item

* `calculate_total()` → returns total price of all items

* `show_items()` → lists all items in order

 

Example:

 

```python

order = Order()

order.add("Shirt", 500)

order.add("Shoes", 1500)

print(order.calculate_total())

```

 

Output:

 

```

Total = 2000

```'''
class Order:
    def __init__(self):
        self.items = {}

    def add(self, item, price):
        self.items[item] = price
        print(f"Added: {item} at {price}")

    def remove(self, item):
        if item in self.items:
            del self.items[item]
            print(f"Removed: {item}")
        else:
            print(f"{item} not found in order")

    def calculate_total(self):
        total = sum(self.items.values())
        return total

    def show_items(self):
        for item, price in self.items.items():
            print(f"{item}: {price}")
order = Order()
order.add("Shirt", 500)
order.add("Shoes", 1500)
print(f"Total = {order.calculate_total()}")
order.show_items()
order.remove("Shirt")
order.show_items()
order.remove("Pants")  
````
